## Chemical space exploration of MEP pathway inhibitors

```python
x_0 = pd.read_csv("../processed/eos4u6p_all_mols.csv")
x_0 = np.array(x_0[x_0.columns[2:]])
x_1 = pd.read_csv("../processed/eos4u6p_mep_activities.csv")
x_1 = np.array(x_1[x_1.columns[2:]])
```

```python
x = np.concatenate([x_0, x_1], axis=0)
from sklearn.decomposition import PCA

reducer = PCA(n_components=4)
reducer.fit(x)
p_0 = reducer.transform(x_0)
p_1 = reducer.transform(x_1)
```

```python
import stylia
import matplotlib.pyplot as plt
fig, axs = stylia.create_figure(1, 1)
ax = axs.next()
ax.scatter(p_0[:, 0], p_0[:, 1], color = "blue", label="All Molecules")
ax.scatter(p_1[:, 0], p_1[:, 1], color = "red", label="MEP Activities")
stylia.label(ax, title="PCA", xlabel="PC1", ylabel="PC2")
stylia.save_figure("../../assets/pca12.png")
```

```python
fig, axs = stylia.create_figure(1, 1)
ax = axs.next()
plt.scatter(p_0[:, 2], p_0[:, 3], color = "blue", label="All Molecules")
plt.scatter(p_1[:, 2], p_1[:, 3], color = "red", label="MEP Activities")
stylia.label(ax, title="PCA", xlabel="PC3", ylabel="PC4")
stylia.save_figure("../../assets/pca34.png")
```

```python
from umap import UMAP

reducer = UMAP(n_components=2)
reducer.fit(x)
p_0 = reducer.transform(x_0)
p_1 = reducer.transform(x_1)
```

```python
fig, axs = stylia.create_figure(1, 1)
ax = axs.next()
ax.scatter(p_0[:, 0], p_0[:, 1], color = "blue", label="All Molecules")
ax.scatter(p_1[:, 0], p_1[:, 1], color = "red", label="MEP Activities")
stylia.label(ax, title="UMAP", xlabel="UMAP1", ylabel="UMAP2")
stylia.save_figure("../../assets/umap.png")
```

## Train a model

```python
x_0 = pd.read_csv("../processed/drugbank_eos4u6p.csv")
x_0 = np.array(x_0[x_0.columns[2:]])
```

```python
from sklearn.linear_model import LogisticRegression

R = []
for i in range(x_0.shape[0]):
    r = x_0[i,:]
    if np.any(np.isnan(r)):
        continue
    R.append(r)
x_0 = np.array(R)

x = np.concatenate([x_1, x_0], axis=0)

y = np.array([1]*x_1.shape[0] + [0]*x_0.shape[0])

model = LogisticRegression(class_weight="balanced")
model.fit(x, y)
```

## Predict for our molecules

```python
x_t = pd.read_csv("../processed/eos4u6p_all_mols.csv")
x_t = np.array(x_t[x_t.columns[2:]])
```

```python
y_hat = model.predict_proba(x_t)[:,1]
```

```python
df = pd.read_csv("../processed/all_mols.csv")
df["mep_prediction"] = y_hat

df.to_csv("../processed/mep_prediction_all_mols.csv", index=False)
```
