```python
import joblib
import pandas as pd
import numpy as np
import zipfile
import os


def save_ersilia_dataframe_as_joblib(df, filename):
    columns = list(df.columns)[2:]
    key = df["key"].tolist()
    input = df["input"].tolist()
    if "eos4wt0" in filename:
        dtype = np.int8
    else:
        dtype = np.float16
    X = np.array(df[columns], dtype=dtype)
    data = (columns, key, input, X)
    joblib.dump(data, filename)
    zip_filename = filename+".zip"
    with zipfile.ZipFile(zip_filename, 'w') as zipf:
        zipf.write(filename, compress_type=zipfile.ZIP_DEFLATED)
    os.remove(filename)


def load_ersilia_dataframe_from_joblib(filename):
    zip_filename = filename+".zip"
    with zipfile.ZipFile(zip_filename, 'r') as zipf:
        with zipf.open(filename) as source_file:
            with open(filename, 'wb') as output_file:
                output_file.write(source_file.read())
    data = joblib.load(filename)
    df_0 = pd.DataFrame({"key": data[1], "input": data[2]})
    df_1 = pd.DataFrame(data[3], columns=data[0])
    os.remove(filename)
    return pd.concat([df_0, df_1], axis=1)


model_id = "eos3804"
df = pd.read_csv("data/{0}_preds.csv".format(model_id))
save_ersilia_dataframe_as_joblib(df, "data/{0}_preds.joblib".format(model_id))
df = load_ersilia_dataframe_from_joblib("data/{0}_preds.joblib".format(model_id))
```
